```python
y=usa['Price']
x=usa[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
sb.distplot(usa['Price'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(x, y)
```

```python
len(Xtrain)/len(x)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
usalr=LinearRegression()
```

```python
usalr.fit(Xtrain,ytrain)
```

```python
usalr.coef_
```

```python
pd.DataFrame(usalr.coef_,index=x.columns)
```

```python
#we have done model summarization
```

```python
pr=usalr.predict(Xtest)
```

```python
r2_score(ytest,pr)
```

```python
mean_absolute_error(ytest,pr)
```
